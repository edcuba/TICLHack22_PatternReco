```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['DejaVu Sans'],'size': 12})
rc('mathtext',**{'default':'regular'})
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
gnn_2p_range = [10, 20, 30, 50, 70, 100]

gnn_nc_2p_clue_F = [0.657833488850464, 0.657833488850464, 0.657833488850464, 0.657833488850464, 0.657833488850464, 0.657833488850464]
gnn_nc_2p_target_F = [0.709613324288395, 0.7814183663627319, 0.8552336878770943, 0.8923024450872056, 0.8965497444336143, 0.901496910968999]
gnn_nc_2p_reco_F = [0.7045536138850059, 0.7699524655659775, 0.8451234492252435, 0.8816242245216624, 0.88365755146593, 0.8875564380144716]

gnn_lp_2p_target_F = [0.8437354538062112, 0.8827909550411878, 0.8949147404605471, 0.896723513091889, 0.8984433591070247, 0.8984433591070247]
gnn_lp_2p_reco_F = [0.8415168841944893, 0.8799705030304544, 0.8912514787270778, 0.8933921037506163, 0.8939779159344413, 0.8887618163668993]
```

```python
fig = plt.figure(figsize=(8, 5))

ax = fig.add_subplot(111)

# ax.plot(gnn_2p_range, (np.array(gnn_lc_2p_target_F) - np.array(gnn_nc_2p_clue_F)), '--s', label="LP target", c="#D55E00")
ax.plot(gnn_2p_range, (np.array(gnn_nc_2p_reco_F) - np.array(gnn_nc_2p_clue_F)), '-o', label="GNN NC", c="#D55E00")
ax.plot(gnn_2p_range, (np.array(gnn_lp_2p_reco_F) - np.array(gnn_nc_2p_clue_F)), '-v', label="GNN LP", c="#56B4E9")

ax.axhline(0.228, label="MLP baseline", c="lightgray", linestyle="--")

ax.legend()
ax.set_xlabel("Neighborhood radius (cm)")
ax.set_ylabel("$\Delta F_{0.5}$")
# ax.set_ylim((-0.1, 0.1))

fig.savefig(f'export/CloseByTwoPion-GNN-radius-plot.png', dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```
